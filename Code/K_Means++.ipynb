```python
def draw_multiple_scatter_plots(datasets, dataset_names, target_name):
    num_plots = len(datasets)
    fig, axes = plt.subplots(1, num_plots, figsize=(4 * num_plots, 4))

    for i, (choosen_dataset, dataset_name) in enumerate(zip(datasets, dataset_names)):
        if target_name != None:
            sns.scatterplot(ax=axes[i], data=choosen_dataset, x='Feature1', y='Feature2', hue=target_name, style=target_name, palette='Set2')
            axes[i].legend(title='Target')
        else:
            sns.scatterplot(ax=axes[i], data=choosen_dataset, x='Feature1', y='Feature2', palette='Set2')
        axes[i].set_title(f'{dataset_name} Scatter Plot')

    plt.tight_layout()
    plt.show()
```

# Step 7: K-means++ Algorithm

```python
def calculate_distance(point1, point2):
    x1 = point1[0]
    y1 = point1[1]
    x2 = point2[0]
    y2 = point2[1]

    return math.sqrt(pow((x2 - x1), 2) + pow((y2 - y1), 2))
```

```python
def k_means_pp(dataset, k, iteration):
    k_list = []
    k_mean = []

    first_centroid = dataset.sample(n=1).iloc[0]
    k_list.append([first_centroid['Feature1'], first_centroid['Feature2']])
    k_mean.append([0, 0, 0])

    for _ in range(1, k):
        k_mean.append([0, 0, 0])
        distances = []
        for index, row in dataset.iterrows():
            distances.append(0)
            for centroid in k_list:
                distances[index] += calculate_distance([row['Feature1'], row['Feature2']], centroid)

        new_centroid_index = distances.index(max(distances))
        k_list.append([dataset['Feature1'][new_centroid_index], dataset['Feature2'][new_centroid_index]])

    for item in range(iteration):
        for index, row in dataset.iterrows():
            distance = []
            for k_number in range(k):
                distance.append(calculate_distance([row['Feature1'], row['Feature2']], k_list[k_number]))
            dataset.loc[index, 'Predicted_Target_' + str(k)] = distance.index(min(distance))
        distance.clear()

        for index, row in dataset.iterrows():
            for index_mean in range(k):
                if row['Predicted_Target_' + str(k)] == index_mean:
                    k_mean[index_mean][0] += row['Feature1']
                    k_mean[index_mean][1] += row['Feature2']
                    k_mean[index_mean][2] += 1

        for index in range(k):
            if k_mean[index][2] != 0:
                k_list[index][0] = k_mean[index][0] / k_mean[index][2]
                k_list[index][1] = k_mean[index][1] / k_mean[index][2]

        for index in range(k):
            k_mean[index] = [0, 0, 0]

    return k_list, dataset
```

```python
k_points = [[], [], []]
for point_index in range(10):
    dataset0_points, datasets[0] = k_means_pp(datasets[0], k = point_index + 1, iteration = 10)
    dataset1_points, datasets[1] = k_means_pp(datasets[1], k = point_index + 1, iteration = 10)
    dataset2_points, datasets[2] = k_means_pp(datasets[2], k = point_index + 1, iteration = 10)
    k_points[0].append(dataset0_points)
    k_points[1].append(dataset1_points)
    k_points[2].append(dataset2_points)
```

# Step 8: Calculating Cost function J(w)

```python
def calculate_cost_function(dataset, k, k_points):
    distance = [0]*k
    numbers = [0]*k
    for index, row in dataset.iterrows():
        which_cluster = round(row['Predicted_Target_' + str(k)])

        k_point_x = k_points[k-1][which_cluster][0]
        k_point_y = k_points[k-1][which_cluster][1]
        point_x = row['Feature1']
        point_y = row['Feature2']

        distance[which_cluster] += calculate_distance([point_x, point_y], [k_point_x, k_point_y])
        numbers[which_cluster] += 1

    WCV = sum(distance)

    golobal_mean_x = 0
    golobal_mean_y = 0
    for index in range(k):
        golobal_mean_x += numbers[index] * k_points[k-1][index][0]
        golobal_mean_y += numbers[index] * k_points[k-1][index][1]
    golobal_mean_x /= sum(numbers)
    golobal_mean_y /= sum(numbers)

    sum_mul = 0
    for index in range(k):
        diference = calculate_distance([k_points[k-1][index][0], k_points[k-1][index][1]], [golobal_mean_x, golobal_mean_y])
        sum_mul += numbers[index]*diference

    if sum_mul == 0:
        BCV = 1
    else:
        BCV = sum_mul

    return BCV/WCV
```

```python
losses = [[], [], []]
for index, dataset in enumerate(datasets):
    for k in range(10):
        losses[index].append(calculate_cost_function(dataset, k+1, k_points[index]))
```

```python
def draw_line_chart(losses, dataset_names):
    num_plots = len(losses)

    fig, axes = plt.subplots(1, num_plots, figsize=(4.5 * num_plots, 4.5))
    sns.set_style("whitegrid")

    if num_plots == 1:
        axes = [axes]

    for i, (loss, name) in enumerate(zip(losses, dataset_names)):
        k_values = list(range(1, len(loss) + 1))
        sns.lineplot(ax=axes[i], x=k_values, y=loss, marker='o', color="blue")

        axes[i].set_title('Elbow Method for ' + name, fontsize=14)
        axes[i].set_xlabel('Number of Clusters (k)', fontsize=12)
        axes[i].set_ylabel('Loss Function (J)', fontsize=12)
        axes[i].set_xticks(k_values)

    plt.tight_layout()
    plt.show()
```
